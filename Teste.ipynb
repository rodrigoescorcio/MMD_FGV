Objetiva-se inicialmente fazer uma análise exploratória dos dados da competição "TalkingData AdTracking Fraud Detection Challenge" do Kaggle (https://www.kaggle.com/c/talkingdata-adtracking-fraud-detection)

Os dados consistem de informações para cada click, como: IP, o aplicativo que estava sendo utilizado, o aparelho utilizado, a versão do sistema operacional do aparelho, o horário do click, se o usuário de fato fez o download e que horas foi feito.

```python
# Importando bibliotecas

import os
import sys
import time
import datetime
import numpy as np
import pandas as pd
import pathlib
import csv
import matplotlib.pyplot as plt
```

```python
csvfile = "teste.csv" #declarar 
```

Primeiramente, devemos importar os dados e definir o nome das colunas.

```python
raw = pd.read_csv(csvfile) #importar dados
raw.columns=['ip','app','device','os','channel','click_data','click_time','clicou','hora_click','dia_click']
```

```python
raw.head() #cabeçalho
```

Pode-se observar que os dados relativos a IP, App, Device, OS e Channel estão codificados. Os *dados click_data* e *click_time* determinam a hora do click, *clicou* determina se a pessoa fez o download e *hora_click* e *dia_click* a hora do download. Porém, na coluna *clicou*, está indicado o dia do click se a pessoa tiver clicado, e 0 se não tiver clicado. Como parte do pré-processamento, vamos colocar 1 se a pessoa tiver clicado, e salvar o dia do click na coluna *dia_click*.

```python
for i in range(len(raw.index)):
    if raw.loc[i,"clicou"] != '0':
        raw.loc[i,"dia_click"] = raw.loc[i,"clicou"]
        raw.loc[i,"clicou"] = '1'
```

Testando, pode-se ver, para um exemplo de click, que as mudanças foram feitas.

```python
print raw.loc[1771,'dia_click']  #teste após a mudança nas colunas
print raw.loc[1771,'hora_click'] 
print raw.loc[1771,'clicou'] 
```

Uma boa maneira de visualizar como os clicks estão distribuídos pelos aplicativos e dispositivos é olhando os histogramas. Pode-se observar que os clicks não-fraudulentos concentram-se em alguns aplicativos e também em alguns dispositivos.

```python
nbins = 100 #Histogramas da nova tabela
plt.hist(esc['app'],bins='auto', density=True, align='mid')
plt.title('Histograma dos apps, em caso de clique.')
plt.ylabel('Frequencia relativa')
plt.show()
plt.hist(esc['device'],bins=nbins, density=True, align='mid')
plt.title('Histograma dos dispositivos, em caso de clique.')
plt.ylabel('Frequencia relativa')
plt.show()
plt.hist(esc['os'],bins='auto', density=True, align='mid')
plt.title('Histograma dos S.O., em caso de clique.')
plt.ylabel('Frequencia relativa')
plt.show()
plt.hist(esc['channel'],bins=nbins, density=True, align='mid')
plt.title('Histograma dos canais, em caso de clique.')
plt.ylabel('Frequencia relativa')
plt.show()


plt.show()
```
