```python
# Importando bibliotecas

import os
import sys
import time
import datetime
import numpy as np
import pandas as pd
import pathlib
import csv
import matplotlib.pyplot as plt
```

```python
csvfile = "teste.csv" #declarar 
```

```python
raw = pd.read_csv(csvfile) #importar dados
raw.columns=['ip','app','device','os','channel','click_data','click_time','clicou','hora_click','dia_click']
```

```python
for i in range(len(raw.index)):
    if raw.loc[i,"clicou"] != '0':
        raw.loc[i,"dia_click"] = raw.loc[i,"clicou"]
        raw.loc[i,"clicou"] = '1'
```

```python
nbins = 100 #Histogramas da nova tabela
plt.hist(esc['app'],bins='auto', density=True, align='mid')
plt.title('Histograma dos apps, em caso de clique.')
plt.ylabel('Frequencia relativa')
plt.show()
plt.hist(esc['device'],bins=nbins, density=True, align='mid')
plt.title('Histograma dos dispositivos, em caso de clique.')
plt.ylabel('Frequencia relativa')
plt.show()
plt.hist(esc['os'],bins='auto', density=True, align='mid')
plt.title('Histograma dos S.O., em caso de clique.')
plt.ylabel('Frequencia relativa')
plt.show()
plt.hist(esc['channel'],bins=nbins, density=True, align='mid')
plt.title('Histograma dos canais, em caso de clique.')
plt.ylabel('Frequencia relativa')
plt.show()


plt.show()
```

```python
from datetime import datetime   
# Importando a biblioteca para poder 
# Passar os dados de string para formato DATA

s = pd.to_datetime(esc['click_data'])
s
```

```python
s.groupby(s.dt.month).count().plot(kind="bar") #Clicks Não-Fraudulentos por Dia
```

```python
r = pd.to_datetime(esc['dia_click']) #Dias de Donwload dos Clicks Não-Fraudulentos 
r.groupby(r.dt.month).count().plot(kind="bar")
```

```python
u = pd.to_datetime(esc['click_time']) #Hora de Acesso dos Clicks Não-Fraudulentos 
u.groupby(u.dt.hour).count().plot(kind="bar")
```

```python
v = pd.to_datetime(esc['hora_click']) #Hora de Download dos Clicks Não-Fraudulentos 
v.groupby(v.dt.hour).count().plot(kind="bar")
```
